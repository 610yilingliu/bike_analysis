```python
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('./cleaned_data/gathered_bike_data.csv')
las, los = df['BIKE_LATITUDE'].tolist(), df['BIKE_LONGITUDE'].tolist()

x = zip(las, los)
```

```python
y_pred = KMeans(n_cluster = 2, random_state = 1).fit_predict(x)
```
