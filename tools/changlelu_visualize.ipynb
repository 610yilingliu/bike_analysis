```python
import folium
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../cleaned_data/fence_position.csv')
```

```python
needs = []
for i in range(len(df)):
    if '长乐路' in df['FENCE_ID'][i]:
        lo = float(df['LONGITUDE'][i])
        la = float(df['LATITUDE'][i])
        needs.append((la, lo))
```

```python
m = folium.Map(location=[24.527352, 118.10321675], zoom_start=13, zoom_control='False',
               attr='AutoNavi')
for point in needs:
    folium.Marker(point).add_to(m)
```
